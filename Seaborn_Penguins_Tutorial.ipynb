```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
#Styles

sns.set_theme(style="darkgrid", context="talk") #white, dark, darkgrid, context=paper, notebook, poster, talk

relational = sns.relplot(data=penguins,
                        height=8,
                        aspect=1.5,
                        markers=["s",".","P","d"])
relational.set(title="Penguins")
relational.set_ylabels("Variables")
relational.set_xlabels("Num. of penguins")


#plt.grid(False)
plt.show()
```

```python
sns.relplot(data=penguins,
            height=8,
            aspect=1.5,
            x="bill_length_mm",
            y="flipper_length_mm")
```

```python
sns.relplot(data=penguins,
            height=8,
            aspect=1.5,
            x="bill_length_mm",
            y="flipper_length_mm",
            hue="body_mass_g",
            s=200)
```

```python
sns.relplot(data=penguins,
            height=8,
            aspect=1.5,
            x="bill_length_mm",
            y="flipper_length_mm",
            hue="species",
            s=200)

sns.relplot(data=penguins,
            height=8,
            aspect=1.5,
            x="bill_length_mm",
            y="flipper_length_mm",
            hue="species",
            row="sex",
            s=200)
```

```python
sns.relplot(data=penguins,
            height=8,
            aspect=1.5,
            x="bill_length_mm",
            y="flipper_length_mm",
            hue="species",
            row="sex",
            size="body_mass_g",
            s=(100,500),
            style="sex")
```

```python
sns.relplot(data=penguins,
            height=6,
            aspect=1,
            x="bill_length_mm",
            y="flipper_length_mm",
            hue="species",
            col="island",
            row="sex",
            size="body_mass_g",
            s=(100,500),
            style="sex")
```

## Lines


```python
sns.relplot(data=penguins,
            height=8,
            aspect=1.5,
            x="species",
            y="body_mass_g",
            kind="line",
            hue="sex",
            style="sex",
            ci="sd")
plt.show()
```

```python
sns.relplot(data=penguins,
            height=8,
            aspect=1.5,
            x="bill_depth_mm",
            y="flipper_length_mm",
            kind="line",
            hue="species",
            style="species",
            ci="sd")
plt.show()
```

```python
sns.relplot(data=penguins,
            height=8,
            aspect=1.5,
            x="bill_depth_mm",
            y="flipper_length_mm",
            kind="line",
            hue="species",
            style="species")
plt.show()
```

## Combination

```python
fig, ax = plt.subplots(1,2, figsize=(15,8))

sns.lineplot(data=penguins,
            x="species",
            y="body_mass_g",
            hue="sex",
            style="sex",
            ci="sd",
            ax=ax[0])

sns.scatterplot(data=penguins,
            x="bill_length_mm",
            y="flipper_length_mm",
            hue="species",
            s=300,
            ax=ax[1])

plt.show()
```
